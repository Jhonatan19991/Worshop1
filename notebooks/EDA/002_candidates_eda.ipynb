# 3. Important information, Outliers? or maybe Patterns?

```python
print(df['Seniority'].value_counts())
```

```python

category_counts = df['Seniority'].value_counts()

plt.barh(category_counts.index, category_counts.values)

plt.xlabel('Count')
plt.ylabel('Seniority')
plt.title('Count of Each Seniority Level')

plt.show()
```

```python
def stadistic_count(num):

    stadistic = num.describe().T
    stadistic['median'] = num.median()
    stadistic = stadistic.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(stadistic)
```

```python
stadistic_count(df.select_dtypes('number'))
```

The data does not seem to have any erroneous scores, as the minimum and maximum values for the scores are 0 and 10, respectively, and there is a maximum of 30 years of experience. Additionally, the numerical values appear to follow a uniform distribution since the mean and median have similar values. This is expected because the data was randomly generated, which typically results in a uniform distribution. Consequently, there are no outliers for the same reason.

```python
numerical = df.select_dtypes('number')
from math import ceil
rows = ceil(numerical.shape[1]/2)

f, ax = plt.subplots(nrows = rows, ncols = 2, figsize = (16, rows * 6))

ax = ax.flat
for i, column in enumerate(numerical.columns):
    df[column].value_counts().plot.barh(ax = ax[i])
    ax[i].set_title(column, fontsize = 12, fontweight = "bold")
```

The distribution of the numerical values indeed appears to follow a uniform distribution, But YOE column does have a different distribution, and very few people have 0 years of experience compared to other years.

# 6. Visualization

In this section, we will explore different data points and ways they can be visualized, in order to assess whether they will be useful for the dashboard or not.

Hires by year (horizontal bar chart)

```python
hired_by_year = df_copy.groupby('year')['hired'].sum().reset_index()

sns.barplot(data=hired_by_year, x='hired', y='year', orient='h')
```

Hires by seniority (bar chart)

```python
hired_by_seniority = df_copy.groupby('Seniority')['hired'].sum().reset_index()

sns.barplot(data=hired_by_seniority, x='Seniority', y='hired')
```

Hires by country over years (USA, Brazil, Colombia, and Ecuador only)(multiline chart)

```python
filteredCountry = df_copy[df_copy['Country'].isin(['United States of America', 'Brazil', 'Colombia', 'Ecuador'])]

hiresByCountryFiltered= filteredCountry.groupby(['year', 'Country'])['hired'].sum().reset_index()

plt.figure(figsize=(12, 8)) 

sns.lineplot(data=hiresByCountryFiltered, x='year', y='hired', hue='Country')

plt.legend(title='Country', fontsize='small', title_fontsize='small')
plt.xlabel('Year', fontsize='medium')
plt.ylabel('Number of Hires', fontsize='medium')
plt.title('Hires by Country Over Years', fontsize='large')

plt.show()
```
